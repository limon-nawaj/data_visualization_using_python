## User Age Distribution by User Type

This violin plot compares the distribution of user ages between different user types. It helps to identify any differences in user age distribution among subscribers and customers.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='user_type', y='member_age')
plt.xlabel('User Type')
plt.ylabel('Age')
plt.title('User Age Distribution by User Type')
plt.show()
```

## Ride Duration by User Type and Gender

This clustered bar chart shows the distribution of ride durations based on user type and gender. It provides insights into how ride durations vary among different user types and genders.

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='user_type', y='duration_min', hue='member_gender', ci=None)
plt.xlabel('User Type')
plt.ylabel('Ride Duration (minutes)')
plt.title('Ride Duration by User Type and Gender')
plt.legend(title='Gender')
plt.show()
```

## Heatmap of User Age and Ride Duration

This heatmap showcases the relationship between user age and ride duration. It provides a visual representation of where ride durations are concentrated for different age groups.

```python
plt.figure(figsize=(10, 6))
heatmap_data = df.groupby(['member_age', 'duration_min']).size().unstack()
sns.heatmap(data=heatmap_data, cmap='YlGnBu', cbar_kws={'label': 'Frequency'})
plt.xlabel('Ride Duration (minutes)')
plt.ylabel('User Age')
plt.title('Heatmap of User Age and Ride Duration')
plt.show()
```

## Pairwise Relationships with PairPlot

This PairPlot provides a matrix of scatter plots for pairs of numerical variables in the dataset. It helps to visualize the relationships between multiple variables at once and identify any potential correlations.

```python
sns.set(style="ticks")
numerical_columns = ['duration_min', 'member_age']
sns.pairplot(df[numerical_columns], diag_kind="kde", markers="o")
plt.suptitle('Pairwise Relationships with PairPlot', y=1.02)
plt.show()
```
